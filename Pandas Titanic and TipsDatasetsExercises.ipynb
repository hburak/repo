```python
#########################################
# Task 6: Check the dtype of embarked variable and convert into categoric variable
#########################################

df["embarked"].dtype
df["embarked"] = df["embarked"].astype("category")
df["embarked"].dtype
df.info()
```

```python
#########################################
# Task 9: Show the women younger than age 30
#########################################

df[(df["age"]<30) & (df["sex"]=="female")].head()
```

```python
#########################################
# Task 10: Show passengers paying more than 500 as fare or younger than age  70.
#########################################

df[(df["fare"] > 500 ) | (df["age"] > 70 )].head()
```

```python
#########################################
# Görev 11: Find null values for each variable 
#########################################

df.isnull().sum()
```

```python
#########################################
# Task 12: drom who variable from dataset.
#########################################

df.drop("who", axis=1, inplace=True)
```

```python
#########################################
# Task 13: replace all the values of the variable deck with the mode of deck.
#########################################

type(df["deck"].mode())
df["deck"].mode()[0]
# since it retuns a pandas series , here we should indicate it as .mode()[0]
```

```python
df["deck"].fillna(df["deck"].mode()[0], inplace=True)
# Here we check if any null value left over.
df["deck"].isnull().sum()
```

```python
#########################################
# Task 14: Fill null values from variable "age" with its median.
#########################################

df["age"].fillna(df["age"].median(),inplace=True)
df.isnull().sum()
```

```python
#########################################
# Task 15: Find sum, count, mean values for survived variable within Pclass and gender cross tabulation.
#########################################

df.groupby(["pclass","sex"]).agg({"survived": ["sum","count","mean"]})
```

```python
#########################################
# Task 16:  Tag the passanger under age 30 as  "1", and tag the rest as"0".
# Appoint these news tages under a new column as "age_flag"
#########################################

def age_30(age):
    if age<30:
        return 1
    else:
        return 0

df["age_flag"] = df["age"].apply(lambda x : age_30(x))

# second solution
df["age_flag"] = df["age"].apply(lambda x: 1 if x<30 else 0)
```

```python
#########################################
# Task 17: Import Tips  dataset from Seaborn.
#########################################

df = sns.load_dataset("tips")
df.head()
df.shape
```

```python
#########################################
# Tasl 18: Calculate sum, min, max and mean for totall_bill variable and group by "time" variable.
#########################################

df.groupby("time").agg({"total_bill": ["sum","min","mean","max"]})
```

```python
#########################################
# Task 19: Calculate min, max and mean for totall_bill variable and group by "time" and "day" variables.
#########################################

df.groupby(["day","time"]).agg({"total_bill": ["sum","min","mean","max"]})
```

```python
#########################################
# Task 20:
#Calculate sum,min, max and mean for totall_bill and tip variables and group it by "day" variable which only having 
#female passangers only in lunch time
#########################################
df[(df["time"] == "Lunch") & (df["sex"] == "Female")].groupby("day").agg({"total_bill": ["sum","min","max","mean"],
                                                                           "tip":  ["sum","min","max","mean"]})
                                                                    
```

```python
#########################################
# Task 21: calculate the mean of orders having size less than "3" and having total_bill more than "10"
#########################################


df.loc[(df["size"] < 3) & (df["total_bill"] >10 ) , "total_bill"].mean() # 17.184965034965035
```

```python
#########################################
# Task 22 : create a new column as total_bill_tip_sum which summing total_bill and tip.
#########################################

df["total_bill_tip_sum"] = df["total_bill"] + df["tip"]
df.head()
```

```python
#########################################
# Task 23: Sort the dataframe according to total_bill_tip_sum variable having top 5 and save it as a new dataframe
#########################################

new_df = df.sort_values("total_bill_tip_sum", ascending=False)[:30]
new_df.shape
```
