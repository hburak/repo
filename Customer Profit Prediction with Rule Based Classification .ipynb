```python
############################################################################################
# TASK 2: Convert satis_checkin_day_diff variable to a new categorical variable called as :EB_Score
############################################################################################

bins = [-1, 7, 30, 90, df["SaleCheckInDayDiff"].max()]
labels = ["Last Minuters", "Potential Planners", "Planners", "Early Bookers"]

df["EB_Score"] = pd.cut(df["SaleCheckInDayDiff"], bins, labels=labels)
df.head(50).to_excel("eb_scorew.xlsx", index=False)
```

```python
df.head()
```

```python
################################################################################################
# TASK 3: Calculate the mean of prices and freqs for the cross tabulations of variables below 
################################################################################################

# City-Concept-EB Score cross tab
df.groupby(by=["SaleCityName", 'ConceptName', "EB_Score" ]).agg({"Price": ["mean", "count"]})
```

```python
# City-Concept-Season cross tab
df.groupby(by=["SaleCityName", "ConceptName", "Seasons"]).agg({"Price": ["mean", "count"]})
```

```python
# City-Concept-CInday cross tab
df.groupby(by=["SaleCityName", "ConceptName", "CInDay"]).agg({"Price": ["mean", "count"]})
```

```python
#############################################
# Task 4: Sort City-Concept-Season cross tab based on PRICE on an ascending order and save it 
#############################################
agg_df = df.groupby(["SaleCityName", "ConceptName", "Seasons"]).agg({"Price": "mean"}).sort_values("Price", ascending=False)
agg_df.head(10)
```

```python
#############################################
# Task 5: Reset the the index in order to solve problem of unnamed variables
#############################################
agg_df.reset_index(inplace=True)

agg_df.head()
```

```python
#############################################
# Task 6: Create a new level based sales variable and save it on the dataset.
#############################################


agg_df['sales_level_based'] = agg_df[["SaleCityName", "ConceptName", "Seasons"]].agg(lambda x: '_'.join(x).upper(), axis=1)
agg_df.head()
```

```python
#############################################
# TASK 7: Create a new price segmentation and save it as SEGMENT variable
#############################################

agg_df["SEGMENT"] = pd.qcut(agg_df["Price"], 4, labels=["D", "C", "B", "A"])
agg_df.head()
```

```python
# Review it as it is grouped by SEGEMENT
agg_df.groupby("SEGMENT").agg({"Price": ["mean", "max", "sum"]})
```
